### TASK 3- Identify countries with largest and smallest age gaps

```python
df_filtered = df[df['Age'] != 'All Ages']

# Calculate age gap for each country (max - min weighted mean)
country_gaps = df_filtered.groupby('Country')['Weighted Mean'].agg(['max', 'min'])
country_gaps['Age_Gap'] = country_gaps['max'] - country_gaps['min']
country_gaps = country_gaps.sort_values('Age_Gap', ascending=False)

# Get top 10 largest and smallest
top_5 = country_gaps.head(10)
bottom_5 = country_gaps.tail(5)
combined = pd.concat([top_5, bottom_5])

# Graph 1: Largest Age Gaps
fig1, ax1 = plt.subplots(figsize=(10, 6))
bars1 = ax1.barh(range(len(top_5)), top_5['Age_Gap'], color='red')
ax1.set_yticks(range(len(top_5)))
ax1.set_yticklabels(top_5.index)
ax1.set_xlabel('Age Gap (Max - Min Weighted Mean)', fontsize=12, fontweight='bold')
ax1.set_title('Countries with LARGEST Age Gaps', fontsize=14, fontweight='bold')
ax1.invert_yaxis()
for i, val in enumerate(top_5['Age_Gap']):
    ax1.text(val + 0.5, i, f'{val:.1f}', va='center', fontweight='bold')
plt.tight_layout()
plt.savefig('largest_age_gaps.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Graph 2: Smallest Age Gaps
fig2, ax2 = plt.subplots(figsize=(10, 6))
bars2 = ax2.barh(range(len(bottom_5)), bottom_5['Age_Gap'], color='green')
ax2.set_yticks(range(len(bottom_5)))
ax2.set_yticklabels(bottom_5.index)
ax2.set_xlabel('Age Gap (Max - Min Weighted Mean)', fontsize=12, fontweight='bold')
ax2.set_title('Countries with SMALLEST Age Gaps', fontsize=14, fontweight='bold')
ax2.invert_yaxis()
for i, val in enumerate(bottom_5['Age_Gap']):
    ax2.text(val + 0.5, i, f'{val:.1f}', va='center', fontweight='bold')
plt.tight_layout()
plt.savefig('smallest_age_gaps.png', dpi=300, bbox_inches='tight')
plt.show()

print("Graphs saved as 'largest_age_gaps.png' and 'smallest_age_gaps.png'")
```

```python


# ===== FILTER DATA =====
# Find question column
question_col = [col for col in df.columns if 'question' in col.lower()][0]

# Filter for the specific renewable energy question
question_text = "How quickly should your country replace coal, oil, and gas with renewable energy"
filtered_df = df[df[question_col].str.contains(question_text, case=False, na=False)].copy()

# Remove 'All Ages' 
filtered_df = filtered_df[filtered_df['Age'] != 'All Ages'].copy()

print(f"Question: {question_text}")
print(f"Total responses: {len(filtered_df)}")
print(f"Countries: {len(filtered_df['Country'].unique())}")
print(f"Age groups: {sorted(filtered_df['Age'].unique())}")
print(f"Response types: {filtered_df['Response'].unique()}")

# ===== CREATE PIVOT TABLE: Country (rows) x Age (columns) =====
print("\n" + "="*70)
print("PIVOT TABLE: Count of Responses by Country and Age")
print("="*70)

# Create pivot table
pivot_table = pd.pivot_table(
    filtered_df, 
    values='Response', 
    index='Country', 
    columns='Age', 
    aggfunc='count',
    fill_value=0
)

# Sort columns by age (if they have numeric patterns)
age_order = sorted(pivot_table.columns)
pivot_table = pivot_table[age_order]

# Add Grand Total column
pivot_table['Grand Total'] = pivot_table.sum(axis=1)

# Sort by Grand Total descending
pivot_table = pivot_table.sort_values('Grand Total', ascending=False)

# Display the table
print(pivot_table)

# Save to CSV
pivot_table.to_csv('pivot_table_country_age.csv')
print("\nâœ“ Pivot table saved to 'pivot_table_country_age.csv'")

# ===== VISUALIZATION 1: Heatmap of Response Counts =====
print("\n" + "="*70)
print("GRAPH 1: Heatmap - Response Count by Country and Age")
print("="*70)

# Create heatmap (without Grand Total column)
heatmap_data = pivot_table.drop('Grand Total', axis=1)

fig, ax = plt.subplots(figsize=(12, max(10, len(heatmap_data) * 0.25)))
sns.heatmap(heatmap_data, annot=True, fmt='g', cmap='YlOrRd', 
            cbar_kws={'label': 'Count of Responses'}, 
            linewidths=0.5, linecolor='white', ax=ax)

ax.set_xlabel('Age Group', fontsize=12, fontweight='bold')
ax.set_ylabel('Country', fontsize=12, fontweight='bold')
ax.set_title(f'Response Count by Country and Age Group\n"{question_text}..."', 
             fontsize=13, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig('graph1_heatmap_counts.png', dpi=300, bbox_inches='tight')
plt.show()

# ===== VISUALIZATION 2: Stacked Bar Chart by Response Type =====
print("\n" + "="*70)
print("GRAPH 2: Response Distribution by Age Group (All Countries Combined)")
print("="*70)

# Create pivot: Age (rows) x Response (columns)
age_response_pivot = pd.pivot_table(
    filtered_df,
    values='Country',  # Just need to count
    index='Age',
    columns='Response',
    aggfunc='count',
    fill_value=0
)

# Define response order
response_order = ['Very quickly', 'Somewhat quickly', 'Slowly', 'Not at all', "Don't know"]
available_responses = [r for r in response_order if r in age_response_pivot.columns]
age_response_pivot = age_response_pivot[available_responses]

# Convert to percentages
age_response_pct = age_response_pivot.div(age_response_pivot.sum(axis=1), axis=0) * 100

# Plot stacked bar
fig, ax = plt.subplots(figsize=(12, 7))
ages = age_response_pct.index
x = np.arange(len(ages))
width = 0.6

colors = ['#27ae60', '#f39c12', '#e67e22', '#c0392b', '#95a5a6']
bottom = np.zeros(len(ages))

for i, response in enumerate(available_responses):
    values = age_response_pct[response].values
    ax.bar(x, values, width, bottom=bottom, label=response, color=colors[i])
    
    # Add percentage labels
    for j, (val, bot) in enumerate(zip(values, bottom)):
        if val > 5:
            ax.text(x[j], bot + val/2, f'{val:.0f}%', 
                   ha='center', va='center', fontsize=10, 
                   fontweight='bold', color='white')
    bottom += values

ax.set_xlabel('Age Group', fontsize=13, fontweight='bold')
ax.set_ylabel('Percentage of Responses', fontsize=13, fontweight='bold')
ax.set_title(f'Response Distribution by Age (All Countries)\n"{question_text}..."', 
             fontsize=13, fontweight='bold', pad=15)
ax.set_xticks(x)
ax.set_xticklabels(ages, rotation=45, ha='right')
ax.set_ylim(0, 100)
ax.legend(title='Response', loc='upper left', bbox_to_anchor=(1, 1))
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('graph2_age_response_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

# Print percentages
print(age_response_pct.round(1))

# ===== VISUALIZATION 3: Top Countries by Age Group =====
print("\n" + "="*70)
print("GRAPH 3: Top 15 Countries by Total Response Count")
print("="*70)

# Get top 15 countries
top15_countries = pivot_table.nlargest(15, 'Grand Total').index

# Filter for top 15
top15_data = pivot_table.loc[top15_countries].drop('Grand Total', axis=1)

# Create grouped bar chart
fig, ax = plt.subplots(figsize=(14, 8))
x = np.arange(len(top15_countries))
width = 0.8 / len(age_order)
colors = ['#3498db', '#2ecc71', '#f39c12', '#e74c3c', '#9b59b6']

for i, age in enumerate(age_order):
    values = top15_data[age].values
    ax.bar(x + i * width, values, width, label=age, color=colors[i % len(colors)])
    
    # Add value labels on bars
    for j, val in enumerate(values):
        if val > 0:
            ax.text(x[j] + i * width, val + 0.5, str(int(val)), 
                   ha='center', va='bottom', fontsize=8)

ax.set_xlabel('Country', fontsize=12, fontweight='bold')
ax.set_ylabel('Count of Responses', fontsize=12, fontweight='bold')
ax.set_title(f'Top 15 Countries by Response Count\n"{question_text}..."', 
             fontsize=13, fontweight='bold', pad=15)
ax.set_xticks(x + width * (len(age_order) - 1) / 2)
ax.set_xticklabels(top15_countries, rotation=45, ha='right')
ax.legend(title='Age Group', loc='upper right')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('graph3_top15_countries_grouped.png', dpi=300, bbox_inches='tight')
plt.show()

# ===== VISUALIZATION 4: Age Gap Analysis (4 Subplots) =====
print("\n" + "="*70)
print("GRAPH 4: Response Distribution by Age (4 Separate Panels)")
print("="*70)

# Define response colors
response_colors = {
    'Very quickly': '#27ae60',
    'Somewhat quickly': '#f39c12',
    'Slowly': '#e67e22',
    'Not at all': '#c0392b',
    "Don't know": '#95a5a6'
}

fig, axes = plt.subplots(2, 2, figsize=(20, 16))
axes = axes.flatten()

for idx, age in enumerate(age_order):
    ax = axes[idx]
    
    # Filter for this age
    age_data = filtered_df[filtered_df['Age'] == age]
    
    # Calculate % by country and response
    country_response_data = []
    for country in age_data['Country'].unique():
        country_subset = age_data[age_data['Country'] == country]
        country_row = {'Country': country}
        
        for response in response_order:
            count = (country_subset['Response'] == response).sum()
            pct = (count / len(country_subset) * 100) if len(country_subset) > 0 else 0
            country_row[response] = pct
        
        country_response_data.append(country_row)
    
    country_df = pd.DataFrame(country_response_data)
    country_df = country_df.sort_values('Very quickly', ascending=True)
    
    # Take top 20
    top_n = min(20, len(country_df))
    country_df = country_df.tail(top_n)
    
    # Plot stacked horizontal bars
    countries = country_df['Country'].values
    y_pos = np.arange(len(countries))
    
    left = np.zeros(len(countries))
    for response in response_order:
        if response in country_df.columns:
            values = country_df[response].values
            ax.barh(y_pos, values, left=left, label=response,
                   color=response_colors[response], height=0.7)
            
            # Add labels
            for i, (val, l) in enumerate(zip(values, left)):
                if val > 8:
                    ax.text(l + val/2, i, f'{val:.0f}%',
                           ha='center', va='center', fontsize=8,
                           fontweight='bold', color='white')
            left += values
    
    ax.set_yticks(y_pos)
    ax.set_yticklabels(countries, fontsize=9)
    ax.set_xlabel('Percentage', fontsize=11, fontweight='bold')
    ax.set_title(f'{age}\n(Top {top_n} countries)', 
                fontsize=12, fontweight='bold', pad=10)
    ax.set_xlim(0, 100)
    ax.grid(axis='x', alpha=0.3)
    
    if idx == 0:
        ax.legend(loc='lower right', fontsize=9)

plt.suptitle(f'Age Gap Analysis Across 4 Age Groups\n"{question_text}..."',
             fontsize=16, fontweight='bold', y=0.995)
plt.tight_layout()
plt.savefig('graph4_age_panels.png', dpi=300, bbox_inches='tight')
plt.show()

# ===== SUMMARY STATISTICS =====
print("\n" + "="*70)
print("SUMMARY: Age Gap Analysis")
print("="*70)

for age in age_order:
    age_data = filtered_df[filtered_df['Age'] == age]
    total = len(age_data)
    very_quickly = (age_data['Response'] == 'Very quickly').sum()
    pct = (very_quickly / total * 100) if total > 0 else 0
    print(f"{age:15s}: {very_quickly:4d}/{total:4d} responses = {pct:5.1f}% 'Very quickly'")

print("="*70)
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import (
    mean_squared_error, mean_absolute_error, r2_score,
    mean_absolute_percentage_error, explained_variance_score
)
import matplotlib.pyplot as plt
import seaborn as sns

# ==================== REGRESSION METRICS ====================
def calculate_regression_metrics(y_true, y_pred, model_name="Model", n_features=1):
    """
    Calculate comprehensive regression evaluation metrics for Weighted Mean prediction
    
    Parameters:
    y_true: array-like, true Weighted Mean values
    y_pred: array-like, predicted Weighted Mean values
    model_name: str, name of the model for display
    n_features: int, number of features used in the model (for Adjusted RÂ²)
    
    Returns:
    dict: Dictionary containing all metrics
    """
    
    # Calculate core metrics
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred) * 100
    
    # Adjusted RÂ² (accounts for number of predictors)
    n = len(y_true)
    p = n_features
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1) if n > p + 1 else r2
    
    # Explained Variance Score
    evs = explained_variance_score(y_true, y_pred)
    
    # Mean Squared Log Error (for non-negative values)
    if np.all(y_true >= 0) and np.all(y_pred >= 0):
        msle = np.mean((np.log1p(y_true) - np.log1p(y_pred)) ** 2)
        rmsle = np.sqrt(msle)
    else:
        msle = None
        rmsle = None
    
    # Max Error
    max_error = np.max(np.abs(y_true - y_pred))
    
    # Mean Error (bias)
    mean_error = np.mean(y_pred - y_true)
    
    # Create results dictionary
    metrics = {
        'Model': model_name,
        'RÂ² Score': r2,
        'Adjusted RÂ²': adj_r2,
        'RMSE': rmse,
        'MSE': mse,
        'MAE': mae,
        'MAPE (%)': mape,
        'RMSLE': rmsle,
        'Explained Variance': evs,
        'Max Error': max_error,
        'Mean Error (Bias)': mean_error
    }
    
    # Print results in a formatted table
    print(f"\n{'='*70}")
    print(f"REGRESSION EVALUATION METRICS - {model_name}")
    print(f"{'='*70}")
    print(f"{'Metric':<30s} {'Value':<20s} {'Interpretation'}")
    print(f"{'-'*70}")
    print(f"{'RÂ² Score':<30s} {r2:>8.4f}          {'Higher is better (max=1.0)'}")
    print(f"{'Adjusted RÂ²':<30s} {adj_r2:>8.4f}          {'Penalizes extra features'}")
    print(f"{'RMSE':<30s} {rmse:>8.4f}          {'Lower is better'}")
    print(f"{'MSE':<30s} {mse:>8.4f}          {'Lower is better'}")
    print(f"{'MAE':<30s} {mae:>8.4f}          {'Lower is better'}")
    print(f"{'MAPE (%)':<30s} {mape:>8.2f}%         {'Lower is better'}")
    
    if rmsle is not None:
        print(f"{'RMSLE':<30s} {rmsle:>8.4f}          {'Lower is better'}")
    else:
        print(f"{'RMSLE':<30s} {'N/A':<8s}          {'(negative values present)'}")
    
    print(f"{'Explained Variance':<30s} {evs:>8.4f}          {'Higher is better (max=1.0)'}")
    print(f"{'Max Error':<30s} {max_error:>8.4f}          {'Largest prediction error'}")
    print(f"{'Mean Error (Bias)':<30s} {mean_error:>8.4f}          {'Should be close to 0'}")
    print(f"{'='*70}\n")
    
    return metrics


# ==================== VISUALIZATION FUNCTIONS ====================
def plot_regression_analysis(y_true, y_pred, model_name="Model"):
    """
    Create comprehensive visualizations for regression results
    """
    fig = plt.figure(figsize=(16, 10))
    
    # 1. Actual vs Predicted
    ax1 = plt.subplot(2, 3, 1)
    ax1.scatter(y_true, y_pred, alpha=0.6, edgecolors='k', s=50)
    ax1.plot([y_true.min(), y_true.max()], 
             [y_true.min(), y_true.max()], 'r--', lw=2, label='Perfect Prediction')
    ax1.set_xlabel('Actual Weighted Mean', fontsize=11, fontweight='bold')
    ax1.set_ylabel('Predicted Weighted Mean', fontsize=11, fontweight='bold')
    ax1.set_title('Actual vs Predicted Values', fontsize=12, fontweight='bold')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # 2. Residuals Plot
    residuals = y_true - y_pred
    ax2 = plt.subplot(2, 3, 2)
    ax2.scatter(y_pred, residuals, alpha=0.6, edgecolors='k', s=50)
    ax2.axhline(y=0, color='r', linestyle='--', lw=2)
    ax2.set_xlabel('Predicted Weighted Mean', fontsize=11, fontweight='bold')
    ax2.set_ylabel('Residuals', fontsize=11, fontweight='bold')
    ax2.set_title('Residual Plot', fontsize=12, fontweight='bold')
    ax2.grid(True, alpha=0.3)
    
    # 3. Residuals Distribution
    ax3 = plt.subplot(2, 3, 3)
    ax3.hist(residuals, bins=30, edgecolor='black', alpha=0.7, color='skyblue')
    ax3.axvline(x=0, color='r', linestyle='--', lw=2)
    ax3.set_xlabel('Residuals', fontsize=11, fontweight='bold')
    ax3.set_ylabel('Frequency', fontsize=11, fontweight='bold')
    ax3.set_title('Residuals Distribution', fontsize=12, fontweight='bold')
    ax3.grid(True, alpha=0.3)
    
    # 4. Q-Q Plot for Residuals
    ax4 = plt.subplot(2, 3, 4)
    from scipy import stats
    stats.probplot(residuals, dist="norm", plot=ax4)
    ax4.set_title('Q-Q Plot (Normality Check)', fontsize=12, fontweight='bold')
    ax4.grid(True, alpha=0.3)
    
    # 5. Error Distribution
    errors = np.abs(residuals)
    ax5 = plt.subplot(2, 3, 5)
    ax5.hist(errors, bins=30, edgecolor='black', alpha=0.7, color='lightcoral')
    ax5.set_xlabel('Absolute Error', fontsize=11, fontweight='bold')
    ax5.set_ylabel('Frequency', fontsize=11, fontweight='bold')
    ax5.set_title('Absolute Error Distribution', fontsize=12, fontweight='bold')
    ax5.grid(True, alpha=0.3)
    
    # 6. Prediction Error by Index
    ax6 = plt.subplot(2, 3, 6)
    ax6.plot(residuals, alpha=0.7, linewidth=0.5)
    ax6.axhline(y=0, color='r', linestyle='--', lw=2)
    ax6.set_xlabel('Sample Index', fontsize=11, fontweight='bold')
    ax6.set_ylabel('Residuals', fontsize=11, fontweight='bold')
    ax6.set_title('Residuals Over Sample Index', fontsize=12, fontweight='bold')
    ax6.grid(True, alpha=0.3)
    
    plt.suptitle(f'{model_name} - Regression Analysis', 
                 fontsize=16, fontweight='bold', y=1.00)
    plt.tight_layout()
    plt.show()


def compare_models(models_dict, y_true):
    """
    Compare multiple models side by side
    
    Parameters:
    models_dict: dict, {'model_name': y_pred_array}
    y_true: array-like, true values
    
    Returns:
    DataFrame with comparison
    """
    results = []
    
    for model_name, y_pred in models_dict.items():
        metrics = calculate_regression_metrics(y_true, y_pred, model_name)
        results.append({
            'Model': model_name,
            'RÂ²': metrics['RÂ² Score'],
            'Adj RÂ²': metrics['Adjusted RÂ²'],
            'RMSE': metrics['RMSE'],
            'MAE': metrics['MAE'],
            'MAPE (%)': metrics['MAPE (%)']
        })
    
    df_comparison = pd.DataFrame(results)
    
    # Sort by RÂ² Score (descending)
    df_comparison = df_comparison.sort_values('RÂ²', ascending=False).reset_index(drop=True)
    
    print("\n" + "="*80)
    print("MODEL COMPARISON TABLE")
    print("="*80)
    print(df_comparison.to_string(index=False))
    print("="*80 + "\n")
    
    return df_comparison


# ==================== EXAMPLE USAGE ====================
if __name__ == "__main__":
    
    print("\nEXAMPLE: Weighted Mean Prediction Evaluation\n")
    
    # Generate sample data (simulating Weighted Mean predictions)
    np.random.seed(42)
    n_samples = 1000
    y_true = np.random.normal(25, 5, n_samples)  # Weighted Mean around 25
    y_pred = y_true + np.random.normal(0, 2, n_samples)  # Add some prediction error
    
    # Calculate metrics for a single model
    metrics = calculate_regression_metrics(
        y_true, y_pred, 
        model_name="Random Forest Regressor",
        n_features=10
    )
    
    # Visualize results
    plot_regression_analysis(y_true, y_pred, "Random Forest Regressor")
    
    # Example: Compare multiple models
    print("\n" + "="*80)
    print("EXAMPLE: Comparing Multiple Models")
    print("="*80)
    
    y_pred_model1 = y_true + np.random.normal(0, 2, n_samples)
    y_pred_model2 = y_true + np.random.normal(0, 3, n_samples)
    y_pred_model3 = y_true + np.random.normal(0, 1.5, n_samples)
    
    models = {
        'Random Forest': y_pred_model1,
        'Linear Regression': y_pred_model2,
        'XGBoost': y_pred_model3
    }
    
    comparison_df = compare_models(models, y_true)
    
    # Visualize comparison
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    metrics_to_plot = ['RÂ²', 'RMSE', 'MAE']
    
    for idx, metric in enumerate(metrics_to_plot):
        axes[idx].bar(comparison_df['Model'], comparison_df[metric], 
                     color=['#1f77b4', '#ff7f0e', '#2ca02c'])
        axes[idx].set_title(f'{metric} Comparison', fontweight='bold')
        axes[idx].set_ylabel(metric)
        axes[idx].tick_params(axis='x', rotation=45)
        axes[idx].grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.show()
```
